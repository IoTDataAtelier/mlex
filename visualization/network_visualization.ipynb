```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
import networkx as nx
G = nx.Graph()
```

```python
path = "/data/pcpe_02.csv"
df = pd.read_csv(path, delimiter=';', decimal=',')
df
```

```python
df = df.values
df
```

```python
mp = {}
cont = 1
G.clear()
for x in df:
    if x[4] not in mp:
        G.add_node(cont)
        mp[x[4]] = cont
        cont+=1
    if x[21] not in mp and not np.isnan(x[21]):
        G.add_node(cont)
        mp[x[21]] = cont
        cont+=1

    if x[4] in mp and x[21] in mp:
        if x[22] == '1':
            G.add_edge(mp[x[4]], mp[x[21]], weight=5)
        else:
            G.add_edge(mp[x[4]], mp[x[21]], weight=0)

print(len(G))
print(G.number_of_edges())
   
```

```python
nx.write_graphml(G, "grafo.graphml")
```
